```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import os
import plotly.express as px
```

```python
mn_df = pd.read_csv("Resources/property_sales.csv")
mn_df.head()
```

```python
# Detecting indices with null values
null = mn_df[mn_df["Gross_Sale"].isna()].index
```

```python
# Removing all the rows with nulls
mn_df.drop(index=null, inplace = True)
```

```python
# Plotting a Histohram for Visualization
Histo_graph = px.histogram(mn_df["Gross_Sale"], histnorm = 'probability')
```

```python
Histo_graph
```

# Observation

From the above diagram, it could be seen that the data is extremely right-skewed. For better visibility, we can impose a restriction on the range of x values.

```python
# Adjusting the range of x values
px.histogram(mn_df["Gross_Sale"], histnorm="probability", nbins = 10000, range_x = [0, 3000000])
```

```python
# Creating random samples
sample1 = mn_df.sample(n = 1000)
sample2 = mn_df.sample(n = 1000)
```

```python
plt.figure(figsize=[10,6])
plt.subplot(1,1,1)

plt.hist(mn_df['Gross_Sale'], bins=100, range=(0,3000000), alpha=0.4, color='green')
plt.hist(sample1['Gross_Sale'], bins=100,range=(0,3000000))
plt.hist(sample2['Gross_Sale'], bins=100, range=(0,3000000), alpha=0.4, color='red')

plt.axvline(mn_df['Gross_Sale'].median(),color='yellow', linewidth=5)
plt.axvline(sample1['Gross_Sale'].median(), linewidth=1, color='red')
plt.axvline(sample2['Gross_Sale'].median(),color='black', linewidth=2, linestyle='dotted')
plt.legend(['Property Sales Median', 'Sample 1 median', 'Sample 2 median','Property Sales','Sample 1', 'Sample 2'])
plt.title("Comparison of 2 random samples")
plt.show()
```

# Findings 

1. There is the observation of similarities between the two samples.

2. From the graph, it indicates that the median of the sample distribution is similar to that of the population.

3. Comparatively, the median value provides more accuracy in the measurement of central tendency than mean.

In conclusion, it can be noted that the similar relationship between the median and distribution values serve as a proof that the two (2) samples were derived from the same population.
