### Data Cleaning 

```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the mouse data and the study results
mouse_data = pd.read_csv("../Resources/Mouse_metadata.csv")
study_data = pd.read_csv("../Resources/Study_results.csv")
```

```python
# Display the mouse data
mouse_data
```

```python
# Display the study data
study_data
```

```python
# Combine the data into a single dataset and display it
combined_data = pd.merge(mouse_data, study_data)
combined_data
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate = combined_data.duplicated(["Mouse ID", "Timepoint"])
duplicate_index = combined_data[duplicate].index
duplicate_mice = combined_data["Mouse ID"].iloc[duplicate_index]
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
combined_data[combined_data[["Mouse ID", "Timepoint"]].duplicated()==True]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID and display it
clean_data = combined_data[combined_data[["Mouse ID", "Timepoint"]].duplicated()==False]
clean_data
```

### Generating the Boxplot

```python
# Determine the final timepoint for each mouse.
final_timepoint = clean_data.groupby("Mouse ID")["Timepoint"].max().reset_index()
# Start by getting the greatest timepoint for each mouse
final_timepoint
```

```python
# Join the newly created `.max()` dataframe to the dataframe from Part 1
joined_data = pd.merge(final_timepoint, combined_data)
joined_data
```

```python
# Create a list with all 10 drug regimens.
drug_list = list(joined_data['Drug Regimen'].unique())

# Create a empty list to fill with the tumor vol data
tumor_data = []

# Isolate (filter) each mice on each drug to collect their tumor volume.
for drug in drug_list:
    

    # Locate the rows which match the drug and get the final tumor volumes of all mice
    final_tumor_vol = joined_data.loc[joined_data["Drug Regimen"] == drug]["Tumor Volume (mm3)"]
    # Append the outcome to the empty list previously created.
    tumor_data.append(final_tumor_vol)
```

```python
# Create a boxplot that visualizes the final tumor volume of all mice in the study across all drug regimens.
drug_regimen_boxplot = plt.boxplot(tumor_data, vert = False, labels = drug_list, flierprops = dict(marker = "*", markerfacecolor = "b", markeredgecolor = "g"))
plt.show()
# Define a custom shape for all outliers in the visualization

# Create horizontal box and whisker plot
```
